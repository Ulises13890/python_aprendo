##Ecuacion de un polinomio en python con clases
Usarenos clases para encapsular el comportamiento de una ecuacion.
$y=$a x^$n


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
class Polinomio:
    def __init__(self, a, n):
        self.n=n
        self.a=a
        self.x_range=np.linspace(-10,10,100)
    def calcular_y(self, x):
        return (self.a*x)**self.n
    def graficar(self):
        plt.plot(self.x_range, self.calcular_y(self.x_range))
```

```python
cuadrado=Polinomio(1,2)
cuadrado.graficar()
```

```python
def lineas_elevadas(x, polinoma:Polinomio):
    y=(polinoma.a*x)**polinoma.n
    return y
def graficarLineas(*polis:Polinomio):
    for polin in polis:
        x=polin.x_range
        plt.plot(x, lineas_elevadas(x, polin))
    plt.show()
    return 0
```

```python
polinomio2=Polinomio(0.5,3)
Polinomio3=Polinomio(2,1)
graficarLineas(cuadrado, polinomio2, Polinomio3)
```

## Converrir lista a CSV
este programa convierte una lista en un archivo csx

```python
frutas = [["Manzana", "Roja", "Dulce"],["Plátano", "Amarillo", "Dulce"], ["Lima", "Verde", "Ácida"]]
```

```python
import pandas as pd
import csv
```

```python
with open("frutas1.csv", mode="w") as frutas_csv:
    escribir_csv=csv.writer(frutas_csv)
    escribir_csv.writerow(["nombre", "color", "sabor"])
    escribir_csv.writerows(frutas)
```

## leer base de datos y convertirla a array

```python
new_array=[]
with open("frutas1.csv", mode="r") as frutas_csv:
    lector_csv=csv.reader(frutas_csv)
    for fila in lector_csv:
        new_array.append(fila)
print(new_array)
```

## Operaciones con JSON
Utilizare la base de datos users de placeholder para manipular el json

```python
import json
import requests
```

```python
url="https://jsonplaceholder.typicode.com/users"
json_usuarios=requests.get(url)
diccionario_usuarios=json.loads(json_usuarios.text)
diccionario_usuarios
```

```python
with open("usuarios.json", mode="w") as json_user:
    json.dump(diccionario_usuarios, json_user)
```
